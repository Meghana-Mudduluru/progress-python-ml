```python
import numpy as np
import matplotlib.pyplot as plt
```

### lineplot
* to visualize data trends over a continuous range,such as time or seq of values
* connects data points for showing changes over time
* to comp mul datasets by overlaying mul lines
#### parameters are -
* marker
* markersize / ms
* markeredgecolor / mec
* markerfacecolor / mfc
* linestyle / ls  dotted,dashed,dashdot,solid,None
* color / c - to set the line color
* linewidth / lw - value is floating number
* xlabel
* ylabel
* title(loc='left') - position of title --- by default it is center

```python
xpoints=np.array([1,2,6,8])
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
# without xpoints
ypoints=np.array([3,8,1,10])
plt.plot(ypoints)
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(xpoints,ypoints,marker='*')
plt.show()
```

#### shortcut string notation - this parameter also called fmt
marker|line|color
* line format can be '-' , ':' , '--' , '-.'

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,'o-.g',ms=10,mec='m',mfc='y')
plt.show()
```

```python
product_id=[1,2,3,4,5]
quantity=[3,15,7,5,10]
plt.plot(product_id,quantity,'o:g',ms=15,mec='m',mfc='y')
plt.xlabel('Product id')
plt.ylabel('Quantity')
plt.title('Product vs Quantity Graph')
plt.show()
```

```python
# using fontdict
product_id=[1,2,3,4,5]
quantity=[3,15,7,5,10]
f1={'color':'blue','size':10}
f2={'color':'m','size':15}
plt.plot(product_id,quantity,'o:g',ms=15,mec='m',mfc='y')
plt.xlabel('Product id',fontdict=f2)
plt.ylabel('Quantity',fontdict=f2)
plt.title('Product vs Quantity Graph',fontdict=f1,loc='left')
plt.grid()
plt.show()
```

### Subplot
* it takes 3 arguments that describes layout of fig
* layout is organised in rows and columns---represented by 1st and 2nd argument
* 3rd arg represents index of current plot

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)
plt.plot(x,y)

#plot2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(2,1,2)
plt.plot(x,y)

plt.suptitle('My Shop')

plt.show()
```

```python
#plot 1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Sales')

#plot2
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('Income')

plt.suptitle('My Shop')

plt.show()
```
