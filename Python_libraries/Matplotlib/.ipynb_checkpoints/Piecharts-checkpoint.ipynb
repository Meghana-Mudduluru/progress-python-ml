### Matplotlib is an open source plotting library
* visualize data using plots and graphs
  

### Pie Charts
#### parameters are -
* labels :  must be an array adds labels to the piechart
* startangle : the default start angle is at x-axis but u can change it (default angle is 0 deg)
* explode : an array that displays how far from the center the wedge is displayed
* shadow : adds a shadow by setting shadow parameter to True
* colors : must be an array with one value for each wedge
* legend : to add list of explanation for each wedge
* legend with header : add title parameter to the legend function
* autopct='%1.1f%%' : displayspercentages on thge slices

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=np.array([35,25,25,10])
plt.pie(y)
#plt.show()
```

```python
# label
y=np.array([35,25,25,10])
labels=['Apples','Mangoes','Strawberry','Guava']
plt.pie(y,labels=labels)
plt.show()
```

```python
# label with startangle
y=np.array([35,25,25,10])
labels=['Apples','Mangoes','Strawberry','Guava']
plt.pie(y,labels=labels,startangle=90)
plt.show()
```

```python
#explode
y=np.array([35,25,25,10])
labels=['Apples','Mangoes','Strawberry','Guava']
explode=[0.1,0,0,0]
plt.pie(y,labels=labels,explode=explode)
plt.show()
```

```python
#shadow
y=np.array([35,25,25,10])
labels=['Apples','Mangoes','Strawberry','Guava']
explode=[0.1,0,0,0]
plt.pie(y,labels=labels,explode=explode,shadow=True)
plt.show()
```

```python
#legend
y=np.array([35,25,25,10])
labels=['Apples','Mangoes','Strawberry','Guava']
plt.pie(y,labels=labels)
plt.legend()
plt.show()
```

```python
#legend with title
y=np.array([35,25,25,10])
labels=['Apples','Mangoes','Strawberry','Guava']
plt.pie(y,labels=labels)
plt.legend(title='Four fruits')
plt.show()
```

```python
# to display percentages
y=[35,25,25,10]
labels=['Apples','Mangoes','Strawberry','Guava']
plt.pie(y,labels=labels,autopct='%1.0f%%')
plt.show()
```
