### Bar graph
* to comp values across categories
* to represent categorical data with rectangular bars
* height of each bar represents val of category
#### parameters are -
* color
* width
* height (for barh)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y) # 1st arg represents categories and 2nd arg represents values
plt.show()
```

```python
# Horizontal bars
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.bar(x,y,color='g',width=0.5) 
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([3,8,1,10])
plt.barh(x,y,color='m',height=0.9) 
plt.show()
```
