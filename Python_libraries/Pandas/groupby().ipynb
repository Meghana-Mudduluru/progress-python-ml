### highest score of virat kohli against which 3 teams

```python
vk1["batsman_runs"].sum()
```

```python
vk1["batsman_runs"].sum().sort_values(ascending=False).head(3)
```

### function to return highest score of batsman against which team

```python
def run_scored(batsman_name):
    vk=delivery["batsman"]==batsman_name
    vk1=delivery[vk].groupby("bowling_team")
    return vk1["batsman_runs"].sum().sort_values(ascending=False).head(3)
```

```python
run_scored("V Kohli")
```

```python
run_scored("DA Warner")
```

```python
run_scored("CH Gayle")
```
