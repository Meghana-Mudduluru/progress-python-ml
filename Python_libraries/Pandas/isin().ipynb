### Find the most destructive death over batsman in the history of IPL


### Strike Rate = (Number of runs/Number of balls) * 100

### Min batsman 200 balls in over 16-20

```python
import numpy as np
import pandas as pd
delivery=pd.read_csv("deliveries.csv")
```

```python
delivery
```

```python
# display only the death overs information
death_over=delivery[delivery["over"]>15]
```

```python
all_batsman=death_over.groupby("batsman")["batsman_runs"].count()
```

```python
x=all_batsman>200
```

```python
all_batsman
```

```python
x
```

```python
all_batsman[x]
```

```python
all_batsman[x].shape
```

```python
#death_over.groupby("batsman")["batsman_runs"].sum()
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
batsman_list
```

### To calculate the strike rate---we should know
#### Runs scored by all these 43 batsman
#### Balls played by these 43 batsman

```python
death_over=death_over[death_over["batsman"].isin(batsman_list)]
```

```python
death_over
```

```python
death_over[death_over["batsman"].isin(batsman_list)].shape
```

```python
final=death_over[death_over["batsman"].isin(batsman_list)]
```

#### to calculate runs

```python
runs=final.groupby("batsman")["batsman_runs"].sum()
```

```python
runs
```

#### to calculate balls

```python
balls=final.groupby("batsman")["batsman_runs"].count()
```

```python
balls
```

#### to calculate Strike Rate

```python
sr=(runs/balls)*100
```

```python
sr
```

```python
sr.head(5)
```

### top 5 batsman with highest strike rate

```python
sr.nlargest()
```

### 5 batsman with least strike rate

```python
sr.nsmallest()
```

```python
sr.nsmallest(3)
```
