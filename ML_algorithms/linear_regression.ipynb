### House pricing dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("BostonHousing.csv")
```

 ### medv is the dependent feature --- the target feature
 ### remaining all are independent features

#### dividing the dataset into dependent and independent features

```python
x=df.iloc[:,:-1]  # independent features  
y=df.iloc[:,-1] # dependent feature
```

#### splitting the dataset into train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

#### chose the model

```python
## Linear Regression
from sklearn.linear_model import LinearRegression
#from sklearn.model_selection import cross_val_score
model=LinearRegression()
```

#### train the model

```python
model.fit(x_train,y_train)
```

#### test the model

```python
predictions=model.predict(x_test)
print(predictions)
```

```python
from sklearn.model_selection import cross_val_score
mse=cross_val_score(model,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
#print(mse)
mean_mse=np.mean(mse)
print(mean_mse)
```

#### evaluation of model

```python
  '''  Continuous values issue : If predictions contains float values, accuracy_score throws an error.
    Fix: Used np.round(predictions).astype(int) to convert them into integer labels.
    Ensuring y_test is integer: Used .astype(int) to avoid mismatches.
    Multiclass Support: Used average='macro' in recall_score and precision_score to handle multiclass cases.

This should work without the ValueError: continuous is not supported error. '''
```

```python
# Convert predictions to discrete labels
predictions = np.round(predictions).astype(int)  # Rounding for binary/multiclass classification

# Ensure y_test is also integer type
y_test = y_test.astype(int)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , recall_score , precision_score
print(accuracy_score(y_test,predictions))
print(recall_score(y_test,predictions,average='macro'))
print(precision_score(y_test,predictions,average='macro'))
print(confusion_matrix(y_test,predictions))
      
```

```python
sns.distplot(predictions,hist=False, color = 'r', label = 'Predicted Values')
sns.distplot(y_test, hist=False, color = 'b', label = 'Actual Values')
plt.legend(loc = "upper left")
plt.show()
```
